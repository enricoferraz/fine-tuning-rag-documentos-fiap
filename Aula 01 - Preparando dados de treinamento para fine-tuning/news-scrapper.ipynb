```python
#Importar bibliotecas
import requests
from bs4 import BeautifulSoup
import re
```

```python
#Definindo a fun??o que retornar? os links

def scrape_cnn_links(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0 Safari/537.36"
    }
    response = requests.get(url, headers=headers, timeout=20)

    # Se a request tiver retorno com sucesso
    if response.status_code == 200:
        # Parsing do conte?do da p?gina
        soup = BeautifulSoup(response.text, 'html.parser')

        valid_links = set()

        # Encontrar os links de not?cias
        for a in soup.find_all('a', href=True):
            href = a['href']

            if href.startswith('http'):
                full_link = href
            elif href.startswith('/'):
                full_link = f"https://edition.cnn.com{href}"
            else:
                continue

            # Filtrar links que parecem ser artigos (padr?o com data no caminho)
            if re.search(r"/\d{4}/\d{2}/\d{2}/", full_link):
                full_link = full_link.split('#')[0].split('?')[0]
                valid_links.add(full_link)

        return sorted(valid_links)
    else:
        return f"Falha ao extrair o conte?do da not?cia: {response.status_code}"
```

```python
# Definindo a URL padr?o
url = "https://edition.cnn.com/world"

# Chama a fun??o de scrap dos dados e salva as URLs em um arquivo
links = scrape_cnn_links(url)
with open('CNN_Links.txt', 'w') as file:
    for link in links:
        file.write(link + '\n')

print(f"Total de links coletados: {len(links)}")
```
